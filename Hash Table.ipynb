```python
class Node:
    def __init__(self, key, data):
        self.next = None
        self.key = key
        self.data = data

class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size
        
    def __len__(self):
        return len([entry for entry in self])
    
    def __setitem__(self, key, data):
        hash_pos = hash(key) % self.size
        curr_node = self.table[hash_pos]
        if curr_node:
            found = False
            while curr_node.next:
                if curr_node.key == key:
                    curr_node.data = data
                    found = True
                curr_node = curr_node.next
            if curr_node.key == key:
                curr_node.data = data
                found = True
            if not found:
                curr_node.next = Node(key, data)
        else:
            self.table[hash_pos] = Node(key, data)
    
    def __getitem__(self, key):
        hash_pos = hash(key) % self.size
        curr_node = self.table[hash_pos]
        if curr_node:
            while curr_node:
                if curr_node.key == key:
                    return curr_node.data
                curr_node = curr_node.next
            raise KeyError(key)
        else:
            raise KeyError(key)
    
    def __iter__(self):
        for node in self.table:
            if node:
                while node:
                    yield node.key, node.data
                    node = node.next
    
    def __contains__(self, key):
        try:
            self[key]
            return True
        except KeyError:
            return False
```

```python
table = HashTable(2)
table['ahah'] = 1
table['aa'] = 2
table['sdsfsdf'] = 1
table['sdsfsdf'] = 4
table['sdsfsdf']
```
