## Mergesort
Write a function that merges two sorted list
then right a function that divides a list in half until there is only one or two elements and return those one or two elements, then merge all the way up.

Time complexity:
* O(n logn) in worst, average, and best case

Space complexity:
* O(n)

NOT in place

IS STABLE

```python
def merge(list1, list2):
    i = j = 0
    result = []
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i += 1
        else:
            result.append(list2[j])
            j += 1
    return result + list1[i:] + list2[j:]

def mergeSort(listy):
    if len(listy) <= 1:
        return listy
    if len(listy) == 2:
        return [min(listy[0], listy[1]), max(listy[0], listy[1])]
    mid = len(listy) // 2
    return merge(mergeSort(listy[:mid]), mergeSort(listy[mid:]))
```

```python
mergeSort([8, 2, 2, 3, 5, 2, 6, 8])
```

## Quicksort
Pick an element to be the pivot, reorder the array forming two partitions such that all elements less than the pivot come before it and all elements greater come after. Recurse.

Ideal strategy: pick the middle element to be the partition and 

Time complexity:
* O(n^2) in worst case
* O(n logn) in average and best case

Space complexity:
* O(n)

NOT STABLE

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = filter(lambda x: x < pivot, arr)
    middle = filter(lambda x: x == pivot, arr)
    right = filter(lambda x: x > pivot, arr)
    return quicksort(left) + middle + quicksort(right)
```

```python
quicksort([8, 2, 2, 3, 5, 2, 6, 8])
```

## Heapsort
Find the maximum element in the array and place it at the end of the array

Time Complexity:
* O(n logn) for best,average,worst

Space Complexity:
* O(1) (in place)

NOT STABLE

```python
def heapify(arr, n, i):
    largest = i
    left = i + 1
    right = i + 2
    if left < n and arr[left] > arr[largest]:
        largest = left
    if right < n and arr[right] > arr[largest]:
        largest = right
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
        
def heapSort(arr):
    for i in xrange(len(arr)-1, -1, -1):
        heapify(arr, len(arr), i)
    for j in xrange(len(arr)-1, -1, -1):
        arr[0], arr[j] = arr[j], arr[0]
        heapify(arr, j, 0)
```
