```python
class Node:
    def __init__(self, key, data):
        self.next = None
        self.prev = None
        self.key = key
        self.data = data

class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size
        
    def __len__(self):
        return self.size        
    
    def __setitem__(self, key, data):
        hash_key = hash(key)
        hash_pos = hash_key % len(self)
        curr_node = self.table[hash_pos]
        if curr_node:
            while curr_node.next:
                curr_node = curr_node.next
            curr_node.next = Node(key, data)
        else:
            self.table[hash_pos] = Node(key, data)
    
    def __getitem__(self, key):
        hash_pos = hash(key) % len(self)
        curr_node = self.table[hash_pos]
        if curr_node:
            while curr_node.next:
                if curr_node.key == key:
                    return curr_node.data
                curr_node = curr_node.next
                
            raise KeyError(key)
        else:
            raise KeyError(key)
    
    def __contains__(self, key):
        try:
            self[key]
        except KeyError:
            return False
        finally:
            return True
```
