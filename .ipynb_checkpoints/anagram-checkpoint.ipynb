```python
def anagramSolution(s1,s2):
    c1 = [0]*26
    c2 = [0]*26

    for char in s1:
        pos = ord(char)-ord('a')
        c1[pos] = c1[pos] + 1

    for char in s2:
        pos = ord(char)-ord('a')
        c2[pos] = c2[pos] + 1

    j = 0
    stillOK = True
    while j<26 and stillOK:
        if c1[j]==c2[j]:
            j = j + 1
        else:
            stillOK = False

    return stillOK

print(anagramSolution('abcd','dcba'))
```
