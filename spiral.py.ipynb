```python
from math import sqrt
import numpy as np
```

```python
#assume n is a perfect square
def spiral(n):
    max_length = len(str(n+1))
    side_length = int(sqrt(n+1))
    N = np.zeros((side_length, side_length), dtype='int')
    x,y = int(side_length/2),int(side_length/2)
    dx,dy = 1,1
    curr_num = 0
    for i in range(1,side_length):
        for n in range(i):
            curr_num += 1
            x += dx
            N[y,x] = curr_num
        dx *= -1
        for n in range(i):
            curr_num += 1
            y += dy 
            N[y,x] = curr_num
        dy *= -1
    for i in range(side_length-1):
        curr_num += 1
        x += dx
        N[y,x] = curr_num
        
    for i in range(side_length):
        for j in range(side_length):
            curr_num = N[i,j]
            padding = " " * (max_length - len(str(curr_num)))
            print padding, curr_num,
        print ""
        
```

```python
spiral(7)
```
