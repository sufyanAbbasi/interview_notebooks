```python
class Node:
    def __init__(self, data):
        self.next = None
        self.prev = None
        self.data = data
        
class LinkedList:
    def __init__(self, head=None):
        self.head = head
        
    def __iter__(self):
        if self.head:
            curr_node = self.head
            while curr_node.next:
                yield curr_node
                curr_node = curr_node.next
            yield curr_node
        else:
            yield self.head
    
    def __str__(self):
        return ' <--> '.join(str(node.data) for node in self) if self.head else '<Empty LinkedList>'
    
    def append_to_tail(self, data):
        curr_node = self.head
        while curr_node.next:
            curr_node = curr_node.next
        tail = Node(data)
        curr_node.next = tail
        tail.prev = curr_node
        
    def delete_node(self, data):
        if self.head:
            if self.head.data == data:
                self.head = self.head.next
            else:
                prev_node = self.head
                curr_node = self.head.next
                while curr_node and curr_node.data != data:
                    prev_node = curr_node
                    curr_node = curr_node.next
                if curr_node:
                    prev_node.next = curr_node.next
                    if curr_node.next:
                        curr_node.next.prev = prev_node
                        
```

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def kth_to_last_node(k, head):
    first_pointer = head
    second_pointer = head
    for i in range(k):
        second_pointer = second_pointer.next
    while second_pointer.next:
        first_pointer = first_pointer.next
        second_pointer = second_pointer.next
    return first_pointer
