```python
def anagramsPrime(str1, str2):
    return len(str1) == len(str2) and reduce(mul, imap(lambda x:primes[ord(x)], str1.lower())) == reduce(mul, imap(lambda x:primes[ord(x)], str2.lower()))   
```

```python
def anagramSolution(s1,s2):
    if len(s1) != len(s2):
        return False
    c1,c2 = [0]*26, [0]*26
    for char1,char2 in izip(s1,s2):
        c1[ord(char1)-ord('a')] += 1
        c2[ord(char2)-ord('a')] += 1
    return c1 == c2
```
