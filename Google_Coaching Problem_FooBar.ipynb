```python
from collections import deque
from random import choice
class IDManager:
    def __init__(self, init=None):
        self.ids = deque([])
        self.indices = {}
        if init:
            for elm in init:
                self.add(elm)
    
    def add(self, i):
        if i not in self.indices:
            self.indices[i] = len(self.ids)
            self.ids.append(i)
    
    def remove(self, i):
        if i in self.indices:
            if len(self.ids) > 2:
                tail = self.ids.pop()
                self.ids[self.indices[i]] = tail
                self.indices[tail] = self.indices.pop(i)
            elif len(self.ids) == 2:
                if i == self.ids[0]:
                    self.indices[self.ids[1]] = 0
                    self.ids.popleft()
                else:
                    self.indices[self.ids[0]] = 0
                    self.ids.pop()
                del self.indices[i]
            else:
                self.ids = deque([])
                self.indices = {}
        else:
            raise Exception('No such element %d' % i)

    def get_random(self):
        return choice(self.ids)
```

```python
from math import sqrt

def sieve_of_eratosthenes(max_len):
    flags = [True]*(max_len + 1)
    flags[0] = flags[1] = False
    prime = 2
    while prime and prime <= sqrt(max_len):
        cross_off(flags, prime)
        prime = get_next_prime(flags, prime)
    return flags
    
def cross_off(flags, prime):
    for i in xrange(prime*prime, len(flags), prime):
        flags[i] = False

def cache_next_prime(func):
    prime_cache = {}
    def cache_prime_helper(flags, prime):
        if not prime:
            return prime
        if prime in prime_cache:
            return prime_cache[prime]
        else:
            next_p = func(flags, prime)
            if next_p:
                prime_cache[prime] = next_p
            return next_p
    return cache_prime_helper

@cache_next_prime
def get_next_prime(flags, prime):
    next_p = prime + 1
    while next_p < len(flags) and not flags[next_p]:
        next_p += 1
    return next_p if next_p < len(flags) and flags[next_p] else None

def answer(n):
    max_len = 10
    factor = 1
    flags = sieve_of_eratosthenes(10)
    prime = 2
    while n > 0:
        n -= len(str(prime))
        prev_prime = prime
        prime = get_next_prime(flags, prime)
        while not prime:
            factor += 1
            flags = sieve_of_eratosthenes(max_len**factor)
            prime = get_next_prime(flags, prev_prime)
    minion_id = str(prime)
    while len(minion_id) < 5:
        prev_prime = prime
        prime = get_next_prime(flags, prime)
        while not prime:
            factor += 1
            flags = sieve_of_eratosthenes(max_len**factor)
            prime = get_next_prime(flags, prev_prime)
        minion_id += str(prime)
    return str(minion_id[:5])
```

```python
def answer(l, t):
    partial_sum = l[0]    
    left_index = right_index = 0
    while right_index < len(l):
        while partial_sum > t and left_index < right_index:
            partial_sum -= l[left_index]
            left_index += 1
        if partial_sum == t:
            return [left_index, right_index]
        
        right_index += 1
        partial_sum += l[right_index] if right_index < len(l) else 0
    
    return [-1, -1]
```

```python
def answer(xs):
    positives = filter(lambda x: x > 0, xs)
    negatives = sorted(filter(lambda x: x < 0, xs))
    zeros = xs.count(0)
    if len(negatives) == 1:
        negative_prod = 0 if zeros else 1 if positives else negatives[0]
    else:
        negative_prod = reduce(lambda x,y: x*y, negatives[:-1] if len(negatives) % 2 else negatives) if negatives else 1
    positive_prod = reduce(lambda x,y: x*y, positives) if positives else 1
    return str(positive_prod * negative_prod if positives or negatives else 0)
```

```python
def memoize(func):
    cache = {}
    def memoize_helper(n, count):
        if n in cache:
            return cache[n] + count
        else:
            result = func(n, count)
            cache[n] = result - count
            return result
    return memoize_helper

@memoize
def answer_helper(n, count):
    if n == 1:
        return count
    if n % 2:
        return min(answer_helper(n + 1, count + 1), answer_helper(n - 1, count + 1))
    else:
        return answer_helper(n/2, count + 1)
    
def answer_v1(n):
    n = int(n)
    if not n:
        return 1
    return answer_helper(n, 0)
```

```python
from math import log

def log_2(n):
    return log(n)/log(2)

def nearest_power(n):
    return 2**int(log_2(n))

def answer_v2(n):
    n = long(n)
    if n == 0:
        return 1
    count = 0
    while n != 1:
        if not n & 1:
            n = n >> 1
        else:
            if n - 1 == nearest_power(n):
                n -= 1
            else:
                left_quotient_odd = (n - 1) >> 1 & 1
                right_quotient_odd = (n + 1) >> 1 & 1
                n = n - 1 if right_quotient_odd else n + 1
        count += 1
    return count
```

```python
from functools import wraps

def memoize(func):
    cache = {}
    @wraps(func)
    def memoizer(*args, **kwargs):
        return cache[tuple(args)] if tuple(args) in cache else cache.setdefault(tuple(args), func(*args, **kwargs))
    return memoizer

def get_sum_up_to(n):
    return n*(n + 1) >> 1

def tallest_stack(n):
    height = 2
    while get_sum_up_to(height) <= n:
        height += 1
    return height - 1

@memoize
def answer_helper(N, curr_sum, n):
    curr_sum += n
    if curr_sum > N:
        return 0
    elif N == curr_sum and n > 1:
        return 1
    else:
        return answer_helper(N, curr_sum, n) + answer_helper(N, curr_sum, n + 1) if n < tallest_stack(N) else 0
    
def answer(n):
    return answer_helper(n, 0, 1)
```

```python
## Implementation of the Edmonds-Karp Algorithm for maximum flow in a flow network

from collections import deque

class Graph:
    def __init__(self, graph):
        self.graph = graph
    
    def bfs(self, source, dest, parent):
        visited = {source}
        queue = deque([source])
        while queue:
            curr_node = queue.popleft()
            for adjacent_node, weight in enumerate(self.graph[curr_node]):
                if weight and adjacent_node not in visited:
                    queue.append(adjacent_node)
                    visited.add(adjacent_node)
                    parent[adjacent_node] = curr_node
        return dest in visited
    
    def edmonds_karp(self, source, dest):
        parent = {}
        max_flow = 0
        # while we can reach the destination from the source in the shortest path
        while self.bfs(source, dest, parent):
            # find the maximum flow found so far
            path_flow = float('Inf')
            curr = dest
            while curr != source:
                # find the minimum flow between the current flow and the parent flow
                path_flow = min(path_flow, self.graph[parent[curr]][curr])
                curr = parent[curr]
            # add the current path flow to the total
            max_flow += path_flow
            
            # adjust weights by path_flow along the path
            curr = dest
            while curr != source:
                prev = parent[curr]
                self.graph[prev][curr] -= path_flow
                self.graph[curr][prev] += path_flow
                curr = parent[curr]
        return max_flow

def answer(entrances, exits, path):
    graph = Graph(path)
    answer = 0
    for entrance in entrances:
        for exit in exits:
            answer += graph.edmonds_karp(entrance, exit)
    return answer
```

Expanding Nebula
================

You've escaped Commander Lambda's exploding space station along with numerous escape pods full of bunnies. But - oh no! - one of the escape pods has flown into a nearby nebula, causing you to lose track of it. You start monitoring the nebula, but unfortunately, just a moment too late to find where the pod went. However, you do find that the gas of the steadily expanding nebula follows a simple pattern, meaning that you should be able to determine the previous state of the gas and narrow down where you might find the pod.

From the scans of the nebula, you have found that it is very flat and distributed in distinct patches, so you can model it as a 2D grid. You find that the current existence of gas in a cell of the grid is determined exactly by its 4 nearby cells, specifically, (1) that cell, (2) the cell below it, (3) the cell to the right of it, and (4) the cell below and to the right of it. If, in the current state, exactly 1 of those 4 cells in the 2x2 block has gas, then it will also have gas in the next state. Otherwise, the cell will be empty in the next state.

For example, let's say the previous state of the grid (p) was:  
.O..  
..O.  
...O  
O...  

To see how this grid will change to become the current grid (c) over the next time step, consider the 2x2 blocks of cells around each cell.  Of the 2x2 block of [p[0][0], p[0][1], p[1][0], p[1][1]], only p[0][1] has gas in it, which means this 2x2 block would become cell c[0][0] with gas in the next time step:  
.O -> O  
..  

Likewise, in the next 2x2 block to the right consisting of [p[0][1], p[0][2], p[1][1], p[1][2]], two of the containing cells have gas, so in the next state of the grid, c[0][1] will NOT have gas:  
O. -> .  
.O  

Following this pattern to its conclusion, from the previous state p, the current state of the grid c will be:  
O.O  
.O.   
O.O   

Note that the resulting output will have 1 fewer row and column, since the bottom and rightmost cells do not have a cell below and to the right of them, respectively.

Write a function answer(g) where g is an array of array of bools saying whether there is gas in each cell (the current scan of the nebula), and return an int with the number of possible previous states that could have resulted in that grid after 1 time step.  For instance, if the function were given the current state c above, it would deduce that the possible previous states were p (given above) as well as its horizontal and vertical reflections, and would return 4. The width of the grid will be between 3 and 50 inclusive, and the height of the grid will be between 3 and 9 inclusive.  The answer will always be less than one billion (10^9).

```python
class GoL:
    def __init__(self, init_state):
        self.state = init_state
        self.ROWS = len(init_state)
        self.COLS = len(init_state[0])
        
    def eval_cell(self, i, j):
        return sum(map(int, iter([self.state[i][j], self.state[i+1][j], self.state[i][j+1], self.state[i+1][j+1]]))) == 1
        
    def next(self):
        new_state = [[None] * (self.COLS - 1) for i in range(self.ROWS - 1)]
        for i in range(self.ROWS - 1):
            for j in range(self.COLS - 1):
                new_state[i][j] = self.eval_cell(i, j)
        self.state = new_state
        return new_state
        
```
